```python
Bz=np.zeros((400,400,200),dtype=float)
```

```python
for line in raw_lines[2:]:
    line=line.strip().split()
    line=[float(x) for x in line]
    x,y,z=int(line[0]*1000),int(line[1]*1000),int(line[2]*1000)
    x+=200
    y+=200
    Bz[x-1,y-1,z-1]=line[5]
```

```python
from matplotlib import pyplot as plt
x=np.arange(0,200,1)
z=np.arange(0,400,1)
X,Z=np.meshgrid(x,z)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(X,Z,np.log10(np.abs(Bz[:,200,:])),cmap='viridis',edgecolor='none')
ax.set_zlim(-5,0)
```
